## 2. Teilmenge (erste 1 Mio. Zeilen) erstellen

```python
df = ords_prods_merge[:1000000]
```

Um Speicherprobleme zu vermeiden, arbeite ich zunächst mit einer Teilmenge von einer Million Zeilen.

## 3. Benutzerdefinierte Funktion price_label definieren

```python
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else:
        return 'Not enough data'
```

Ich definiere eine Funktion, die Produkte anhand ihres Preises in Preisbereiche einteilt.

## 4. Funktion anwenden & neue Spalte erstellen

```python
df.head()
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

Die Funktion wird zeilenweise angewendet und erzeugt die neue Spalte price_range.

## 5. Verteilung prüfen

```python
df['price_range'].value_counts()
```

Ich prüfe die Häufigkeit der erzeugten Kategorien

## 7. Preislabels mit loc() erstellen (empfohlene Methode)

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

## 8. Verteilung der loc‑Labels prüfen

```python
df['price_range_loc'].value_counts()
```

Ich überprüfe, ob beide Methoden zu denselben Ergebnissen führen.

## 9. loc‑Methode auf den gesamten Dataframe anwenden

```python
ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range'] = 'High-range product'
ords_prods_merge.loc[(ords_prods_merge['prices'] <= 15) & (ords_prods_merge['prices'] > 5), 'price_range'] = 'Mid-range product'
ords_prods_merge.loc[ords_prods_merge['prices'] <= 5, 'price_range'] = 'Low-range product'
```
