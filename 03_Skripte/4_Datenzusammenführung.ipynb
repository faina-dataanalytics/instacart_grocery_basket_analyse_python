## 1. DataFrames importieren

```python
# Bibliotheken importieren
import pandas as pd
import numpy as np
import os

# Pfade definieren
base_path = r"C:\Users\faink\Desktop\Achievement_4\02_Daten"
raw_path = os.path.join(base_path, "Rohe_Daten")
prepared_path = os.path.join(base_path, "Prepared_Daten")

# Produkte importieren (aus Rohe_Daten)
df_ords_prior = pd.read_csv(os.path.join(raw_path, 'orders_products_prior.csv'))
df_ords = pd.read_csv(os.path.join(prepared_path, 'orders_checked.csv'))
```

## 4. Merging von df_ords und df_ords_prior

```python
df_merged = df_ords.merge(df_ords_prior, on='order_id', how='inner')
```

Ich führe df_ords und df_ords_prior über die gemeinsame Spalte order_id zusammen.
Da beide Dataframes vollständige Übereinstimmungen enthalten sollten, verwende ich einen inner join.

## 5. Prüfung auf vollständige Übereinstimmung (indicator=True)

```python
df_merged_check = df_ords.merge(df_ords_prior, on='order_id', how='inner', indicator=True)
df_merged_check['_merge'].value_counts()
```

Mit dem Argument indicator=True überprüfe ich, ob alle Keys in beiden Dataframes vorhanden sind.
Wenn ausschließlich der Wert both erscheint, liegt eine vollständige Übereinstimmung vor.

## 6. Export des kombinierten Dataframes im Pickle-Format

```python
df_merged.to_pickle(r'02_Daten/Prepared_Daten/df_merged.pkl')
```

Ich exportiere den zusammengeführten Dataframe im Pickle-Format, um Speicherplatz zu sparen und die Ladezeit zu verkürzen.
